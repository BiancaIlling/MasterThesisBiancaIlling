This notebook includes an initial analysis of the optimization results. First, the new consumption patterns of one household under dynamic energy prices are compared to those in the empirical data, both using a volumetric grid tariff. Then, the impacts on the distribution grid are analyzed in the same scenario for the same time span.

### Requirements

Load the necessary packages for the power flow calculation and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandapower.networks as nw
import pandapower.plotting as plot
import pandapower as pp
from numpy import pi
```

### Load Data

```python
d_before = pd.read_csv('load_before_shift.csv', index_col=0)
ev_hourly = pd.read_csv('Hourly_EV_Charging.csv', index_col=0)
hp_before = pd.read_csv('hp_before_shift.csv', index_col=0)
d_after = pd.read_csv('load_after_shift.csv', index_col=0)
ev_after = pd.read_csv('ev_after_shift_4kappa.csv', index_col=0)
hp_after = pd.read_csv('hp_after_shift_4kappa.csv', index_col=0)
unique_garage = pd.read_csv('Unique_Garage.csv')
price = pd.read_csv('Hourly_Flexible_Prices.csv')
```

Calculate the EV load before shifting. In the DataFrame ev_hourly, each user is separated. Now, the sum of all loads of the users per garage is needed.

```python
frame= np.zeros((len(hp_before.index),len(hp_before.columns)))
ev_before = pd.DataFrame(frame, index=hp_before.index, columns=hp_before.columns)
for i in range(len(hp_before.columns)):
    selected_columns = ev_hourly.filter(regex=f'^{unique_garage.iloc[i,0]}-', axis=1)
    for j in range (len(hp_before)):
        load=0
        for k in range(len(selected_columns.columns)//2):
            load += selected_columns.iloc[j,k*2]
        ev_before.iloc[j,i] = load
```

Define plotting settings:

### Visualize Timestep 9075 with highest line loading in the days 14th&15th January

```python
#Find timestep with maximum line loading within the interval
ll_a.iloc[t_start:t_end,:].max(axis=1).idxmax() 
```

```python
#Create a Kerber Dorfnetz and convert the consumption per hour in mWh
net_ll = nw.create_kerber_dorfnetz()
d_after_mWh = d_after/1000
```

```python
#randomly sample each household 2 or 3 times
import random
random.seed(1)

l = [i for i in range(22)]
l.extend([i for i in range(22)])
l.extend(random.sample(range(22), 13))

# Shuffle the list
random.shuffle(l)

loadlist = d_after.columns.tolist()
```

```python
for load_idx in net_ll.load.index:
    # Assign a value to the load based on its index
    new_load_value = d_after_mWh.loc[9075, loadlist[l[load_idx]]]
    
    # Update the load value in the network
    net_ll.load.at[load_idx, 'p_mw'] = new_load_value
```

```python
#Plot to create artificial geodata
plot.simple_plot(net_ll, plot_loads=True, load_size=0.7, bus_size=0.50, trafo_size=1.0, ext_grid_size=1.0)
```

```python
pp.runpp(net_ll, max_iteration=20)
```

```python
#Define custom colormap for visualization
cmap_list=[(20, "green"), (60, "yellow"), (100, "red")]
cmap, norm = plot.cmap_continuous(cmap_list)
```

```python
#Visualize status of the grid
lc = plot.create_line_collection(net_ll, net_ll.line.index, zorder=1, cmap=cmap, norm=norm, linewidths=2)
bc = plot.create_bus_collection(net_ll, size=.05, zorder=10, color="k")
sl = plot.create_ext_grid_collection(net_ll, size=0.4, orientation=3*pi/2, linewidths=2.)
t21 = plot.create_trafo_collection(net_ll, linewidths=2.)
lo = plot.create_load_collection(net_ll, size=0.18)
plot.draw_collections([lc,bc,sl,t21, lo], figsize=(8,6))
plt.savefig('TimestepOfMaxLL1415012024.png')
```
